```python
#identifying unique invoice number to make the purchase list
invoice_no=df.InvoiceNo.unique()
len(invoice_no)
```

```python
#making the items list on each purchase
sales=[]
invoice=[]
k=0
for i in invoice_no[:2500]:
    invoice.append(i)
    df1 =list(np.array(short.loc[short['InvoiceNo'] == i].product_id))
    print(df1)
    sales.append(df1)
    #items_list.append(items)
```

```python
# dataframe before the association pattern mining
df2={'invoice_no':invoice,'items':sales}
data=pd.DataFrame(df2)
data
```

```python
# applying apriori algorithm
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_array=te.fit_transform(sales)
```

```python
te_array
```

```python
dfx=pd.DataFrame(te_array,columns=[te.columns_])
dfx
```

```python
# 280 assosciations found
min_sup=0.02
freq_items=apriori(dfx,min_sup,use_colnames=True)
freq_items
```

```python
# final output
freq_items['length']=freq_items['itemsets'].apply(lambda x:len(x))
freq_items
```
